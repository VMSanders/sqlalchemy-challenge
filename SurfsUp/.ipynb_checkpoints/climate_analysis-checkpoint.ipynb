```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# Create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# Reflect an existing database into a new model. 
# Definitely grabbed this code from the reflection_solution in 3/27/23 activities.
Base = automap_base()
Base.prepare(autoload_with=engine)
Base.classes.keys()
```

```python
# Assign classes to their own variables
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create a session
session = Session(engine)
```

```python
# Display the first row of each table because I have no idea what they look like.
first_measurement = session.query(Measurement).first()
first_measurement.__dict__
```

```python
first_station = session.query(Station).first()
first_station.__dict__
```

```python
engine.execute('SELECT * FROM Measurement LIMIT 5').fetchall()
# That's in a different order than I thought it would be.
```

```python
# Find the most recent date in the precipitation measurement dataset.
session.query(Measurement.date).order_by(Measurement.date.desc()).first()
```

```python
# So I want all the data from 08/23/2016 onward? Let's grab all those dates.
last_year_dates = session.query(Measurement.date).\
    filter(Measurement.date > '2016-08-23').\
    order_by(Measurement.date).all()
```

```python
# We can have a few different observations on the same date, I see.
last_year_dates[0:5]
```

```python
# Let's query the last 12 months worth of precipitation data.
last_year_prcp = session.query(Measurement.prcp).\
    filter(Measurement.date > '2016-08-23').\
    order_by(Measurement.date).all()
```

```python
# What does that precipitation data look like?
last_year_prcp[0:5]
# Not a list of numbers. This will continue to surprise me.
```

```python
# Let's pull the precipitation and dates into numbers and strings respectively so that I'll know how to work with them.
last_year_data=[(last_year_dates[i][0],last_year_prcp[i][0]) for i in range(len(last_year_dates))]
```

```python
# Throw the dates and precipitation measurements into a pandas dataframe. That's way easier for me to deal with.
last_year_df = pd.DataFrame(last_year_data,columns=["Date","Precipitation"])
last_year_df.fillna(0,inplace=True)
last_year_df.head()
```

```python
# plot the date vs precipitation
last_year_df.plot(x='Date', y='Precipitation', rot=90)
plt.xlabel("Date")
plt.ylabel("Inches")
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data

last_year_df[['Precipitation']].describe()
```

```python
# Design a query to calculate the total number stations in the dataset
print(f"Number of stations: {len(engine.execute('SELECT * FROM Station').fetchall())}")
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
# I think I'll have to join the 2 tables.
```

```python
# extract the station names to a list
station_list = [entry[1] for entry in engine.execute('SELECT * FROM Station').fetchall()]
station_list
```

```python
# count the number of measurements for each station
observation_counts = [session.query(Measurement.tobs).\
    filter(Measurement.station ==name).count() for name in station_list]
observation_counts
```

```python
# put the stations and observation counts in a pandas dataframe because I like the way it looks
station_count_df = pd.DataFrame(zip(station_list,observation_counts),columns=["Station","Num Observations"])
station_count_df.sort_values('Num Observations',ascending=False,inplace=True)  
station_count_df
```
